```python
tasks = []
completed = []

def show_progress():
    total = len(tasks) + len(completed)
    
    if total == 0:
        print("No tasks yet.")
        return

    progress = int((len(completed) / total) * 20)
    bar = "█" * progress + "-" * (20 - progress)

    percentage = (len(completed) / total) * 100
    print(f"\nProgress: [{bar}] {percentage:.1f}%")

while True:
    print("\n===== TO-DO LIST MENU =====")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Complete Task")
    print("4. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        task = input("Enter new task: ")
        tasks.append(task)
        print("Task added!")

    elif choice == "2":
        print("\nPending Tasks:")
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")

        print("\nCompleted Tasks:")
        for task in completed:
            print(f"✔ {task}")

        show_progress()

    elif choice == "3":
        if not tasks:
            print("No tasks to complete.")
            continue

        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")

        num = int(input("Enter task number completed: "))
        finished = tasks.pop(num - 1)
        completed.append(finished)
        print("Task completed!")

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid option!")
```
